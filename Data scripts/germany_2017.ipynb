```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import pickle
import random
import time
```

```python
# Get the website containing a list with all constituencies
result = requests.get("https://www.bundeswahlleiter.de/en/bundestagswahlen/2017/ergebnisse/bund-99.html")
result
```

```python
# Convert the respons to bs
content = result.content
soup = BeautifulSoup(content, "lxml")
soup
```

```python
district = soup.findAll("ul", {"class": "linklist"})
district = district[1]
region_urls = []
for element in district:
    try:
        a_elements = element.findAll('a')
        for a in a_elements:
            region_urls.append("https://www.bundeswahlleiter.de/en/bundestagswahlen/2017/ergebnisse/"+a['href'])
    except:
        None
region_urls
```

```python
constituency_urls = []
for region_url in region_urls:
    print("new region")
    result = requests.get(region_url)
    content = result.content
    soup = BeautifulSoup(content, "lxml")
    
    url_lists = soup.find("div", {"class": "flex-row cols-2 cols-md-1"})
    for l in url_lists:
    #     print(l)
        for element in l:
            try:
                a_elements = element.findAll('a')
                for a in a_elements:
                    constituency_urls.append(region_url[:-5]+"/"+a['href'].split('/', 1)[-1])
                    print(a['href'].split('/', 1)[-1])

            except:
                None
constituency_urls
```

```python
election_results = []
counter = 0
import random
import time
for constituency_url in constituency_urls:
    timeDelay = random.randrange(0, 1)
    time.sleep(timeDelay)
    print("in constituency", counter, "of", len(constituency_urls))
    result = requests.get(constituency_url)
    content = result.content
    soup = BeautifulSoup(content, "lxml")
    data_htmls = soup.findAll("span", {"class": "data-figure__data"})
    total_results = []
    winner = True
    constituency_name = soup.findAll("div", {"class":"dropdown"})[1]
    for elements in constituency_name:
        try:
            print(elements.text)
            constituency_name = elements.text
            constituency_name = constituency_name.split(":", 1)[-1]
            constituency_name = constituency_name.lstrip()
            break
        except:
            None
    for i in ["Germany", "2017",constituency_name]:
        total_results.append(i)
            
    for data_html in data_htmls:
        sub_result = []
        for element in data_html:
            try:
                sub_result.append(element.text)
            except:
                sub_result.append(element)
                None
        name = sub_result[0].strip()
        party = sub_result[2]
        percentage = sub_result[4].strip()[:4]   
        for i in [name, party, percentage, winner]:
            total_results.append(i)
        winner = False
    election_results.append(total_results)
    counter += 1
    
    
```

```python
final_df = pd.DataFrame(election_results)
final_df
```

```python
final_df.to_csv('election_results_germany_2017_2.csv')
```

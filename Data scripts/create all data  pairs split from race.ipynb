```python
import pandas as pd
```

```python
df = pd.read_csv('data/election_results_uk_2015_2017_2019_finished.csv')
df = df[[df.columns[4],df.columns[5], df.columns[7],df.columns[8],df.columns[9],df.columns[11] ]]
df['winner_name'] = df[df.columns[0]]+ ' '+df[df.columns[1]]
df['winner'] = df[df.columns[2]]
df['loser_name'] = df[df.columns[3]]+ ' '+df[df.columns[4]]
df['loser'] = df[df.columns[5]]
df = df[['winner_name', 'winner', 'loser_name', 'loser']]
df_uk = df
df_uk.head()
```

```python
df_germany = pd.read_csv('data/election_results_germany_2017.csv')
df_germany = df_germany[[df_germany.columns[4], df_germany.columns[5], df_germany.columns[7], df_germany.columns[8], df_germany.columns[9], df_germany.columns[11]]]
df_germany.columns = ['winner_name', 'party_win', 'winner','loser_name', 'party_los', 'loser']
df_germany
df_germany = df_germany.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df_germany
df_germany['winner_name'] = df_germany['winner_name']+" "+df_germany['party_win']
df_germany['loser_name'] = df_germany['loser_name']+" "+df_germany['party_los']
df_germany = df_germany[['winner_name', 'winner', 'loser_name', 'loser']]
df_germany
df_germany.head()
```

```python
df = pd.read_csv('data/france_2017.csv')
winner_names = []
loser_names = []
for winner_name, loser_name in zip(df[df.columns[4]].str.split(), df[df.columns[8]].str.split()):
    winner_names.append(' '.join(winner_name[1:]))
    loser_names.append(' '.join(loser_name[1:]))
    
df[df.columns[4]] = winner_names
df[df.columns[8]] = loser_names 
df = df[[df.columns[4], df.columns[5], df.columns[7],df.columns[8], df.columns[9], df.columns[11]]]
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df.columns = ['winner_name', 'party_win', 'winner','loser_name', 'party_los', 'loser']

df['winner_name'] = df['winner_name']+' candidat'
df['loser_name'] = df['loser_name']+' candidat'
df_france = df
df_france = df_france[['winner_name', 'winner', 'loser_name','loser']]
df_france.head()
```

```python
df = pd.read_csv('data/canada_2015_2019.csv')
df = df[[df.columns[4], df.columns[5], df.columns[7], df.columns[8], df.columns[9], df.columns[11]]]
df.columns = ['winner_name', 'party_win', 'winner','loser_name', 'party_los', 'loser']
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df['winner_name'] = df['winner_name']+" "+df['party_win']
df['loser_name'] = df['loser_name']+" "+df['party_los']# df_canada = all_candidates[['filename', 'winner']]
df_canada = df[['winner_name', 'winner', 'loser_name', 'loser']]
df_canada.head()
```

```python
import re
df = pd.read_csv('data/usa_2018.csv')
df = df[[df.columns[2], df.columns[3], df.columns[4],df.columns[6], df.columns[7], df.columns[8]]]
df.columns = ['winner_name', 'party_win', 'winner', 'loser_name', 'party_los', 'loser']
p = re.compile(r'[^\w\s]+')
df['winner_name'] = [p.sub('', x) for x in df['winner_name'].tolist()]
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df['winner_name'] = df['winner_name']+" "+df['party_win']
df['loser_name'] = df['loser_name']+" "+df['party_los']
df_usa = df[['winner_name', 'winner', 'loser_name', 'loser']]
df_usa.head()
```

```python
all_results = pd.concat([df_usa, df_canada, df_france, df_uk, df_germany])
all_results
import unidecode
def decode(x):
    return unidecode.unidecode(x)
all_results['loser_name'] = all_results['loser_name'].astype(str)
all_results[all_results.columns[0]] = all_results[all_results.columns[0]].apply(decode)
all_results[all_results.columns[2]] = all_results[all_results.columns[2]].apply(decode)
all_results = all_results.applymap(lambda x: x.strip() if isinstance(x, str) else x)
print(len(all_results))
all_results.drop_duplicates(inplace = True)
print(len(all_results))

all_results.reset_index(inplace = True, drop = True)
all_results.head()
```

```python
test_set_len = int(len(all_results) * 0.1)
train_len = int(len(all_results) - test_set_len )
all_results = all_results.sample(frac=1).reset_index(drop=True)
all_results
test_df= all_results[-test_set_len:]
train_df = all_results[:train_len]
train_df
val_len = int(len(train_df) * 0.05)
train_len = int(len(train_df) - val_len )
val_df = train_df[-val_len:]
train_df = train_df[:train_len]
```

```python
# import unidecode
def decode(x):
    return unidecode.unidecode(x)
train_df[train_df.columns[0]] = train_df[train_df.columns[0]].apply(decode)
train_df[train_df.columns[2]] = train_df[train_df.columns[2]].apply(decode)
train_df = train_df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
# train_df.reset_index(inplace = True, drop = True)

train_df['winner_name'] = 'allimgs_face/'+train_df['winner_name']
train_df['loser_name'] = 'allimgs_face/'+train_df['loser_name']


test_df[test_df.columns[0]] = test_df[test_df.columns[0]].apply(decode)
test_df[test_df.columns[2]] = test_df[test_df.columns[2]].apply(decode)
test_df = test_df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
test_df['winner_name'] = 'allimgs_face/'+test_df['winner_name']
test_df['loser_name'] = 'allimgs_face/'+test_df['loser_name']
# test_df.reset_index(inplace = True, drop = True)



val_df[val_df.columns[0]] = val_df[val_df.columns[0]].apply(decode)
val_df[val_df.columns[2]] = val_df[val_df.columns[2]].apply(decode)
val_df = val_df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
val_df['winner_name'] = 'allimgs_face/'+val_df['winner_name']
val_df['loser_name'] = 'allimgs_face/'+val_df['loser_name']
# val_df.reset_index(inplace = True, drop = True)

val_df
```

```python
# # Remove races with missing candidates
import os
def remove_missing_races(df):
    counter = 0
    clean_data = []
    for folder_win, folder_los in zip(df['winner_name'].values, df['loser_name'].values):
        del_row = False
        if not os.path.isdir(folder_win):
#             print(folder_win, 'NOT FOUND')
            del_row = True
        if not os.path.isdir(folder_los):
#             print(folder_los, 'NOT FOUND')
            del_row = True
            
        if not del_row:
            clean_data.append([folder_win, True, folder_los, False])
            
            
    return pd.DataFrame(clean_data, columns = ['winner_name', 'winner', 'loser_name', 'loser'])
            
            
            

train_df_clean = remove_missing_races(train_df)
test_df_clean = remove_missing_races(test_df)
val_df_clean = remove_missing_races(val_df)
```

```python
# # create images pairs
import os
import glob
def create_image_pairs(df):
    all_img_data = []
    counter = 0
    for folder_win, winner, folder_los, loser in zip(df['winner_name'], df['winner'],df['loser_name'], df['loser']):

        if folder_win not in ['allimgs_face/Candidate Votes', 'allimgs_face/. MARIE SARA candidat']:
            if os.path.isdir(folder_win):
                files_win = [f for f in glob.glob(folder_win + "**/*", recursive=True)]
            else:
                print('folder win not found',folder_win)
        if folder_los not in ['allimgs_face/Candidate Votes', 'allimgs_face/. MARIE SARA candidat']:    
            if os.path.isdir(folder_los):
                files_los = [f for f in glob.glob(folder_los + "**/*", recursive=True)]
            else:
                print('folder los not found',folder_los)
            for file_win in files_win:
                for file_los in files_los:
                    all_img_data.append([file_win, winner, file_los, loser])
                    counter +=1 
    final_df = pd.DataFrame(all_img_data, columns = ['winner_name', 'winner', 'loser_name', 'loser'])
    final_df.drop_duplicates(inplace = True)
    return final_df

print("TRAIN")
train_df_imgs_pairs = create_image_pairs(train_df_clean)
print("TEST")
test_df_imgs_pairs = create_image_pairs(test_df_clean)
print("VAL")
val_df_imgs_pairs = create_image_pairs(val_df_clean)
print("DONE")

    
```

```python
# # Remove small images
from PIL import Image
import numpy as np

def remove_small_imgs(all_img_df):
    no_small_imgs = []
    for win_img_name, los_img_name in zip(all_img_df[all_img_df.columns[0]].values,all_img_df[all_img_df.columns[2]].values):
        if '.webp' not in win_img_name and '.webp' not in los_img_name and not win_img_name.endswith('.txt') and not los_img_name.endswith('.txt'):
            im_win = Image.open(win_img_name)
            im_win_width, im_win_height = im_win.size

            im_los = Image.open(los_img_name)
            im_los_width, im_los_height = im_los.size

            if im_win_width < 100 or im_win_height < 100 or im_los_width < 100 or im_los_height < 100:
                None
            else:
                no_small_imgs.append([win_img_name, True,los_img_name, False])


    all_img_df_no_small = pd.DataFrame(no_small_imgs, columns = ['winner_name', 'winner', 'loser_name', 'loser'])
    return all_img_df_no_small

train_df_imgs_pairs_nosmall = remove_small_imgs(train_df_imgs_pairs)
test_df_imgs_pairs_nosmall = remove_small_imgs(test_df_imgs_pairs)
val_df_imgs_pairs_nosmall = remove_small_imgs(val_df_imgs_pairs)
train_df_imgs_pairs_nosmall
```

```python
winners = train_df_imgs_pairs_nosmall[[train_df_imgs_pairs_nosmall.columns[0]]]
winners.drop_duplicates(inplace = True)
losers = train_df_imgs_pairs_nosmall[[train_df_imgs_pairs_nosmall.columns[2]]]
losers.drop_duplicates(inplace = True)
print("loser percentage", len(losers) / (len(losers) + len(winners)))
```

```python
import pandas as pd
from PIL import Image
import cv2
import numpy as np
```

```python
df_train = pd.read_csv('data/train_data_clean.csv')
df_train = df_train[~df_train[df_train.columns[1]].str.contains('.wepb')]
df_test = pd.read_csv('data/test_data_clean.csv')
df_test = df_test[~df_test[df_test.columns[1]].str.contains('.wepb')]
print(len(df_test) + len(df_train))
path_test = df_train[df_train.columns[1]]
path_train = df_test[df_test.columns[1]]

paths = []
for path in path_test:
    if '.webp' not in path:
        paths.append(path)
for path in path_train:
    if '.webp' not in path:
        paths.append(path)
print(len(paths))
        
    
# df.head()
# print(df[df.columns[1]][802])
```

```python
import os
import dlib
face_cascade = cv2.CascadeClassifier('xml_files/haarcascade_frontalface_default.xml')
dnnFaceDetector = dlib.cnn_face_detection_model_v1('xml_files/mmod_human_face_detector.dat')
```

```python
i = 0
for path in paths[3053:]:
    new_path = 'temp/'+ path.split('/')[1]+"/"+path.split('/')[2]
    new_folder = 'temp/'+ path.split('/')[1]  
    print(new_path)
    try:
        os.mkdir(new_folder)
    except:
        print("Dir", new_folder, "already exists")
#     print("OPENING IMG")
        
    image = Image.open(path).convert('RGB')
    img = np.array(image)

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    face = face_cascade.detectMultiScale(gray, 1.1, 5)    
    if type(face) is not tuple:
        x,y,w,h = face[0][0],face[0][1],face[0][2],face[0][3]
    else:
        print("Haar cascade has not found a face")
        face = dnnFaceDetector(gray, 1)
        face = face[0]
        x = face.rect.left()
        y = face.rect.top()
        w = face.rect.right() - x
        h = face.rect.bottom() - y
        
        
    crop_img = img[y:y+h, x:x+w]
    crop_image = Image.fromarray(crop_img, 'RGB')
    
    crop_image.save(new_path)
    i +=1
    print(i)
```

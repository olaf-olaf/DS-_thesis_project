```python
# Create a dataframe with all candidates from UK
```

```python
import pandas as pd
```

```python
uk_one = pd.read_csv('data/election_results_uk_2015_2017_2019_finished.csv')
uk_one.head()
```

```python
uk_two = pd.read_csv('data/uk2010results.csv')
uk_two.head()
```

```python
uk_all = pd.concat([uk_one,uk_two])
len(uk_all)
```

```python
uk_winners = uk_all[[uk_all.columns[3],uk_all.columns[4],uk_all.columns[5] ,uk_all.columns[7]]]
uk_winners.columns = ['constituency', 'candidate', 'party', 'winner']
uk_winners
```

```python
uk_losers = uk_all[[uk_all.columns[3],uk_all.columns[8],uk_all.columns[9] ,uk_all.columns[11]]]
uk_losers.columns = ['constituency', 'candidate', 'party', 'winner']
uk_losers.head()
```

```python
uk_all_candidates = pd.concat([uk_winners, uk_losers])
uk_all_candidates['candidate'] = uk_all_candidates['candidate'].astype(str)
uk_all_candidates['party'] = uk_all_candidates['party'].astype(str)
uk_all_candidates = uk_all_candidates.applymap(lambda x: x.strip() if isinstance(x, str) else x)
uk_all_candidates.drop_duplicates(inplace = True, subset = ['candidate', 'party'])
uk_all_candidates
```

```python
uk_all_candidates
```

```python
uk_all_candidates.to_csv('data/uk_all_candidates.csv')
```
